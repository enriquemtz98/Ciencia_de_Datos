## Puntos extra
- Investigar los siguientes comandos: ?df.dropna() ?df.astype
dropna(): comando incluido en la libreria pandas, nos ayuda a eliminar los datos faltantes. Parametros:
        dropna(self, axis=0, how='any', thresh=None, subset=None, inplace=False)
            axis: elegimos igualandolo a 0 si queremos eliminar la fila que lo contiene e igualandolo a 1 si queremos eliminar             la columna que lo contiene
            how: elegimos 'any' si queremos eliminar toda la fila/columna que contenga aunque sea un NA, elegimos 'all' si                 queremos que se elimine toda la fila/columna solamente si hay puros NA
            thresh: ponemos un limite de datos para mantener la columna, por ejemplo, lo igualamos a dos para que solo se                   elimine si la fila/columna no contiene 2 datos o más
            subset: si estamos eliminando filas, con este comando le indicamos las columnas que son cruciales eliminar si                   cuentan con NAs y viceversa
            inplace: mantiene los valores del dataframe en la misma variable 
astype(): cambia un objeto entero de pandas de tipo. Parametros:
        astype(self: ~FrameOrSeries, dtype, copy: bool = True, errors: str = 'raise') 
            dtype: utiliza un tipo de python para modificar todo el dataframe a ese tipo. Tambien se puede especificar por                 columna
            copy: si es TRUE, te regresa una copia, si es FALSE, los cambios que generemos a nuestro nuevo dataframe afectaran             al original
            errors: si es 'raise' el nuevo dataframe saldra aunque se tenga errores, si es 'ignore', no se le realizaran                   cambios al dataframe en caso de que tenga errores
            
- Realizar una tarea de preprocesamiento no mencionada en esta práctica.

```python
from sklearn.impute import SimpleImputer
basedatosfinal.astype(str).dtypes
basedatosfinal=basedatosfinal.replace('[?ÿ]','',regex=True)
basedatosfinal=basedatosfinal.replace('[,132,"""475""",530,640,73058679...	]','',regex=True)
X = basedatosfinal.iloc[:, :].values
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 
imputer = imputer.fit(X[:, 1:])
X[:, 1:] = imputer.transform(X[:, 1:])
dataframeminado=pd.DataFrame(X,columns=['Titulo','Duración','Calificación'])
dataframeminado
```
